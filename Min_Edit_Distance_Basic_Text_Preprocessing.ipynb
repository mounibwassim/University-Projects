**Tokenization**

A word (Token) is the minimal unit that a machine can understand and process. So any text string cannot be further processed without going through tokenization. Tokenization is the process of splitting the raw string into meaningful tokens. The complexity of tokenization varies according to the need of the NLP application, and the complexity of the language itself.

**Stemming**

Stemming, in literal terms, is the process of cutting down the branches of a tree to its stem. So effectively, with the use of some basic rules, any token can be cut down to its stem.

Stemming is more of a crude rule-based process by which we want to
club together different variations of the token.

Let's initialize the word list before stemming as:

*plurals = ['generously', 'having', 'flies', 'dies', 'ate', 'denied', 'agreed', 'owned', 'humbled', 'sized', 'meeting', 'stating', 'itemization', 'traditional', 'reference', 'plotted']*

```python
plurals = ['generously', 'having', 'flies', 'dies', 'ate', 'denied', 'agreed', 'owned', 'humbled', 'sized', 'meeting', 'stating', 'itemization', 'traditional', 'reference', 'plotted']
```

**LancasterStemmer**

First, try LancasterStemmer:

```python
from nltk.stem.lancaster import LancasterStemmer

lst = LancasterStemmer()

root = [lst.stem(w) for w in plurals]

print("Before:", plurals)
print("After:", root)
```

**PorterStemmer**

Let's try PorterStemmer:

```python
from nltk.stem.porter import PorterStemmer

pst = PorterStemmer()

root = [pst.stem(w) for w in plurals]

print("Before:", plurals)
print("After:", root)
```

**SnowballStemmer**

Now let's try SnowballStemmer:

```python
from nltk.stem.snowball import SnowballStemmer

sbst = SnowballStemmer("english")

root = [sbst.stem(w) for w in plurals]

print("Before:", plurals)
print("After:", root)
```

**Lemmatization**

While stemming is used most of the time for its simplicity, there are cases of complex language or complex NLP tasks where it's necessary to use lemmatization instead.

Lemmatization is a more robust and methodical way of combining grammatical variations to the root of a word. Lemmatization uses context and part of speech to determine the inflected form of the word and applies different normalization rules for each part of speech to get the root word (lemma).


**WordNetLemmatizer**

Here, WordNetLemmatizer is using wordnet, which takes a word and searches wordnet, a semantic dictionary. It also uses a morph analysis to cut to the root and search for the specific lemma (variation of the word).

Let's perform lemmatization with WordNetLemmatizer. By default, the pos argument in WordNetLemmatizer.lemmatize() is noun ('n').

```python
import nltk
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')

wnl = WordNetLemmatizer()
root = [wnl.lemmatize(w) for w in plurals]

print("\n")
print("Before:", plurals)
print("After:", root)
```

WordNetLemmatizer allows us to specify the part of speech to get better lemmas. We specify the pos argument as verb ('v').

```python
root = [wnl.lemmatize(w, pos='v') for w in plurals]

print("Before:", plurals)
print("After:", root)
```

**Stop word removal**

Stop word removal is one of the most commonly used preprocessing steps across different NLP applications. The idea is simply removing the words that occur commonly across all the documents in the corpus. Typically, articles and pronouns are generally classified as stop words. These words have no significance in some of the NLP tasks like information retrieval and classification, which means these words are not very discriminative. NLTK comes with a pre-built list of stop words for around 22 languages.


```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stoplist = set(stopwords.words('english'))

print(stoplist)
```

```python
from nltk.tokenize import word_tokenize

s2 = "This is a sample sentence, showing off the stop words filtration."

word_tokens = word_tokenize(s2)
filtered_sentence = [w for w in word_tokens if not w.lower() in stoplist]

print(word_tokens)
print(filtered_sentence)
```

**Spelling correction**

It is not a necessary to use a spellchecker for all NLP applications, but some use cases require you to use a basic spellcheck. We can create a very basic spellchecker by just using a dictionary lookup.


```python
from nltk.metrics.distance import edit_distance
```

Now, we download the ‘words’ resource (which contains correct spellings of words) and import it through nltk.corpus and assign it to correct_words.

Next, we define the list of incorrect_words for which we need the correct spellings.

Then we run a loop for each word in the incorrect words list in which we calculate the Edit distance of the incorrect word with each correct spelling word having the same initial letter.

We then sort them in ascending order so the shortest distance is on top and extract the word corresponding to it and print it.

```python
nltk.download('words')
from nltk.corpus import words
correct_words = words.words()

incorrect_words = ['happpy', 'azmaing', 'intelliengt']

for word in incorrect_words:
	temp = [(edit_distance(word, w),w) for w in correct_words if w[0]==word[0]]  # store as (edit distance, word)
	print(word, "==>", sorted(temp)[0][1])
```

Sentence or paragraph comparison is useful in applications like plagiarism detection.

```python
sent1 = "It might help to re-install Python if possible."
sent2 = "It might help to re-install C++ if possible."
sent3 = "It can help to install Python again if possible."

ed_sent_1_2 = edit_distance(sent1, sent2)
ed_sent_1_3 = edit_distance(sent1, sent3)

print('Edit Distance between sent1 and sent2: ', ed_sent_1_2)

print('Edit Distance between sent1 and sent3: ', ed_sent_1_3)
```
